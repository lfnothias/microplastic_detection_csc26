# CorSeaCare detector training (Colab GPU)

Runtime → Change runtime type → GPU. Trains the YOLO11 detector on a YOLO-det
dataset (e.g. DeepParticle prepared with `corseacare.data.prepare_deepparticle`).
Ultralytics YOLO is AGPL-3.0.

```python
!pip -q install ultralytics
from google.colab import drive; drive.mount('/content/drive')
```

```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")              # AGPL-3.0
model.train(data="/content/drive/MyDrive/corseacare/data.yaml",
            epochs=100, imgsz=640, batch=16, project="corseacare", name="poc")
```

```python
metrics = model.val()                   # prints mAP50, mAP50-95
print(metrics.box.map, metrics.box.map50)
model.export(format="onnx")             # optional, for portable local inference
```
